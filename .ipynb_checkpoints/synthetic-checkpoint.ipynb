```python
import matplotlib.pyplot as plt
import numpy as np
import bruges
import seaborn as sns
from matplotlib import colors
from PIL import Image

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
channel = np.asarray(Image.open('punta_real.png'))

x = channel.shape[0] 
y = channel.shape[1] * channel.shape[2] 
 
channel.resize((x,y)) # a 2D array 

plt.figure(figsize = (10, 10))
im = plt.imshow(channel, aspect = 3)
# plt.colorbar(shrink = 0.4)
```

```python
flattened = channel.flatten()
import collections
counted = collections.Counter(flattened)
counted = {k: v for k, v in sorted(counted.items(), key=lambda item: item[1], reverse = True)}
keys = np.array(list(counted.keys()))
```

```python
keys = sorted(keys[0:5])
keys
```

```python
channel = np.where((channel > 0) & (channel <= (keys[1])), 1, channel)
channel = np.where((channel > keys[1]) & (channel <= keys[2]), 2, channel)
channel = np.where((channel > keys[2]) & (channel <= keys[3]), 3, channel)
channel = np.where((channel > keys[3]) & (channel <= keys[4]), 4, channel)
channel = np.where((channel > keys[4]), 5, channel)
np.unique(channel)
```

```python
cmap = colors.ListedColormap(['k', 'white', 'grey', 'tab:red', 'tab:orange'])

plt.figure(figsize = (5, 5))
im = plt.imshow(channel, aspect = 4, cmap = cmap)

N = 5
cbar = plt.colorbar(ticks=range(5), cmap = cmap)
plt.clim(-0.5, N - 0.5)
```

```python
zero = [2700, 2750] # Vp, rho
one = [3100, 3150] # Vp, rho
two = [2200, 2250] # Vp, rho
three = [2800, 3000] # Vp, rho
four = [3100, 3150] # Vp, rho

rocks = np.array([zero, one, two, three, four])
```

```python
earth = rocks[channel]
```

```python
earth.shape
```

```python
imp = np.apply_along_axis(np.product, -1, earth)
```

```python
rc = np.diff(imp, axis=0)

plt.imshow(rc, cmap = 'Greys', aspect = 4)
plt.xlabel('Length (m)')
plt.ylabel('Depth (m)')

plt.show()
```

```python
plt.figure(figsize = (4, 2))
w = bruges.filters.ricker(duration = 0.1, dt = 0.001, f = 50)
plt.plot(w)
plt.show()
```

```python
synth = np.apply_along_axis(lambda t: np.convolve(t, w, mode='same'), axis=0, arr=rc)

plt.figure(figsize = (7, 7))
plt.imshow(synth, cmap = "RdBu", aspect = 4, vmin = -1e6, vmax = 1e6)
plt.xlabel('Length (m)', fontsize = 12)
plt.ylabel('Depth (m)', fontsize = 12)

plt.colorbar(shrink = 0.4)
plt.show()
```
